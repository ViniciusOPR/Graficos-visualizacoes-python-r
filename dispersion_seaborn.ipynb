Gráficos de Dispersão com Seaborn

```python
import pandas as pd
import seaborn as srn
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv('co2.csv')
base.head()
```

```python
#Gráfico de dispersão utilizando os atributos conc e uptake, agrupamento pelo type
srn.scatterplot(data=base, x='conc', y='uptake', hue='Type')
plt.show()
```

```python
#Seleção de registros específicos da base de dados (Quebec e Mississipi)
q = base.loc[base['Type'] == 'Quebec']
m = base.loc[base['Type'] == 'Mississippi']
```

```python
#Subgráfico (1 linha e duas colunas) mostrando gráficos sobre cada região
plt.figure(figsize=(10, 5)) # Ajuste o tamanho se necessário
plt.subplot(1, 2, 1)
srn.scatterplot(x=q.conc, y=q.uptake)
plt.title('Quebec')
srn.scatterplot(data=base, x='conc', y='uptake', hue='Type')
plt.subplot(1, 2, 2)
srn.scatterplot(x=m.conc, y=m.uptake)
plt.title('Mississippi')
plt.tight_layout()
plt.show()
```

```python
#Refrigerado e não refrigerado
ch = base.loc[base['Treatment'] == 'chilled']
nc = base.loc[base['Treatment'] == 'nonchilled']
```

```python
#Gráfico somente com 'chilled' e 'nonchilled'
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
srn.scatterplot(x=ch.conc, y=ch.uptake)
plt.title('Chilled')
plt.subplot(1, 2, 2)
srn.scatterplot(x=nc.conc, y=nc.uptake)
plt.title('Non chilled')
plt.tight_layout()
plt.show()
```
