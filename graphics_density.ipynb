Gráficos de Densidade

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base = pd.read_csv('trees.csv')
base.head()
```

```python
# Histograma com 10 divisões (bins) e somente para o primeiro atributo da base de dados
plt.hist(base.iloc[:,1], bins=6)
```

```python
#Histograma com a linha de distribuição de frequência com 6 divisões (bins), kde = linha de densidade
sns.histplot(base.iloc[:,1], kde=False, bins=6, color='blue').set(title="Árvores")
```

```python
#Densidade, método kdeplot para densidade
sns.kdeplot(base.iloc[:,1], color='orange').set(title='Árvores')
```

```python
#Densidade e Histograma
sns.histplot(base.iloc[:,1], kde=True, bins=6, color='orange').set(title='Árvores')
```
