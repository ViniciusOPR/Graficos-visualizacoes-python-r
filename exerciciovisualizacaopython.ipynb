```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as srn
```

```python
base = pd.read_csv('dados_python.csv', sep=';')
base
```

```python
srn.boxplot(data=base)
```

```python
srn.histplot(data=base)
```

```python
#PIB & ValorEmpenho
pb = base.loc[base['PIB'] == 'PIB']
ve = base.loc[base['VALOREMPENHO'] == 'Valores de Empenho']
```

```python
#Agrupamentos - PIB & VALOREMPENHO
plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1)
plt.scatter(base.PIB, base.VALOREMPENHO)
plt.subplot(2, 2, 2)
plt.hist(base.PIB)
plt.subplot(2, 2, 3)
plt.hist(base.VALOREMPENHO)
```

```python
agrupado_pib = base.sort_values('PIB').head(10)
agrupado_pib = agrupado_pib.iloc[:,1:3]
agrupado_pib
agrupado_pib.plot.bar(x='MUNICIPIO',y='PIB', color = 'lightblue')
```

```python
agrupado_empenho = base.sort_values('VALOREMPENHO').head(10)
agrupado_empenho = agrupado_empenho.iloc[:,[1,3]]
agrupado_empenho
agrupado_empenho.plot.bar(x='MUNICIPIO',y='VALOREMPENHO', color = 'orange')
```

```python
ag = base.groupby(['PIB'])['MUNICIPIO'].sum()
ag
```

```python
#10 Maiores PIBS
dez_maiores_pibs = base.nlargest(10,"PIB")
dez_maiores_pibs.plot.bar(x='MUNICIPIO',y='PIB', color = 'darkgreen')
```

```python
#10 Maiores Empenhos
dez_maiores_empenhos = base.nlargest(10,"VALOREMPENHO")
dez_maiores_empenhos.plot.bar(x='MUNICIPIO',y='VALOREMPENHO', color = 'purple')
```
